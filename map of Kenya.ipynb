```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
```

```python
counties = gpd.read_file("shapes/County.shp")
```

```python
counties.head()
```

```python
busia_county = counties[counties.COUNTY=='Busia']
busia_county
```

```python
busia_county['geometry']
```

```python
counties.plot()
ax = plt.gca()
busia_county['geometry'].plot(ax=ax, color='red')
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.title('Kenya with Busia County in Red')
plt.savefig('Figures/Kenya_map.png')
plt.show()
```
